```python
import pandas as pd
```

## Leitura de conjunto de dados

```python
df_arrecadacao = pd.read_csv('https://gist.githubusercontent.com/omadson/770934cb0df9e2d2c3e8db031f1139a5/raw/4d181efc035b327d9d66256915951384517c603e/arrecadacao.csv')
df_arrecadacao.head()
```

```python
new_df = df_arrecadacao.assign(arrecadacao_milhao = df_arrecadacao.arrecadacao / 1000000)
new_df
```

## Visualização de dados

```python
ax = sns.barplot(data=new_df, y='arrecadacao_milhao', x='titulo')
```

```python
ordem_alfabetica = new_df.sort_values(by='arrecadacao_milhao', ascending=False)
ax = sns.barplot(data=ordem_alfabetica, y='arrecadacao_milhao', x='titulo')
```

```python
sns.set_style('whitegrid')

ax = sns.barplot(data=ordem_alfabetica, y='arrecadacao_milhao', x='titulo')
```

```python
ordem_alfabetica = new_df.sort_values(by='arrecadacao_milhao', ascending=False)
plt.figure(figsize=(5, 2))
ax = sns.barplot(data=ordem_alfabetica, x='arrecadacao_milhao', y='titulo', color = 'tomato')

ax.bar_label(
    ax.containers[0],
    fmt='%.1f',
    label_type='center',
    color='white'
)

plt.xlabel('Arrecadação em Milhões de dólares')
plt.ylabel('Filme')
plt.title('Arrecadação de filmes')

ax.set(frame_on=False)
ax.axvline(0, color='k', clip_on=False)
plt.show()
```
