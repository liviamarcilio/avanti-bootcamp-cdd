## Carregar bibliotecas

```python
import seaborn as sns
```

```python
from matplotlib import pyplot as plt
```

## Conjunto de dados

```python
df_gorjetas = sns.load_dataset('tips')
df_gorjetas.head()
```

## Visualização dos dados

```python
ax = sns.boxplot(x='day', y='total_bill', data=df_gorjetas)
sns.set_style("whitegrid")
```

```python
ax = sns.boxplot(x='day', y='total_bill', data=df_gorjetas, color="cornflowerblue")
sns.set_style("whitegrid")
plt.figure(figsize=(6, 3))
```

```python
ax = sns.boxplot(x='day', y='total_bill', data=df_gorjetas, color="cornflowerblue")
sns.set_style("whitegrid")
plt.xlabel('Dia da semana')
plt.ylabel('Conta (USD)')
plt.title('Distribuição de gorjetas ao longo da semana')
plt.show()
```
